```python
def randomization(n):
    """
    Arg:
      n - an integer
    Returns:
      A - a randomly-generated nx1 Numpy array.
    """
    
    A = np.random.random([n,1])
    
    return A
```

```python
def operations(h, w):
    """
    Takes two inputs, h and w, and makes two Numpy arrays A and B of size
    h x w, and returns A, B, and s, the sum of A and B.

    Arg:
      h - an integer describing the height of A and B
      w - an integer describing the width of A and B
    Returns (in this order):
      A - a randomly-generated h x w Numpy array.
      B - a randomly-generated h x w Numpy array.
      s - the sum of A and B.
    """
    A = np.random.random([h,w])
    B = np.random.random([h,w])
    s = A + B

    return A, B, s
```

```python
def norm(A, B):
    """
    Takes two Numpy column arrays, A and B, and returns the L2 norm of their
    sum.

    Arg:
      A - a Numpy array
      B - a Numpy array
    Returns:
      s - the L2 norm of A+B.
    """
    s = np.linalg.norm(A+B)

    return s
```

```python
def neural_network(inputs, weights):
    """
     Takes an input vector and runs it through a 1-layer neural network
     with a given weight matrix and returns the output.

     Arg:
       inputs - 2 x 1 NumPy array
       weights - 2 x 1 NumPy array
     Returns (in this order):
       out - a 1 x 1 NumPy array, representing the output of the neural network
    """
    inputs=np.transpose(inputs)
    dot_prod = np.matmul(inputs,weights)
    output = np.tanh(dot_prod)
    
    return output
```

```python
# condiciones iniciales
theta = np.array([[0.5,0.5],[6,7],[1,4]])
# datos
x = np.resize(np.array([-1,0,4,5,6]),(5,1))

for i in range(8):
    last_theta = np.copy(theta) 
    # E-Step
        # peso_j * Norm(x, theta) 
    p_j_N = theta[0] * ss.norm.pdf(x , loc=theta[1] , scale=np.sqrt(theta[2]))
        #  P(j|i)
    p_j_i =  p_j_N/np.resize(np.sum(p_j_N, axis=1), (5,1))

    # M-Step
        # hat n
    hn = np.sum(p_j_i,axis=0)

        # peso_j
    theta[0] = hn / len(x)
        
        # hat mean
    theta[1] = np.sum( p_j_i * x , axis=0) / hn 

        # square sigma 
    theta[2] = np.sum( p_j_i * (x-theta[1])**2, axis=0) / (hn*1)    # d = 1 
    
    print('Difference = ' , theta-last_theta )

```

## Q-value iteration homework 5 

```python
V=np.zeros(5)
g= 0.6
T = np.zeros((6,6,2))
R = np.zeros((6,6,2))
Q = np.zeros((1,6,2))
```

```python
T[1:6,0:5,0]=np.identity(5)
T[0,0,0]=1

T[[0,4,5],[0,4,5],1]=1
T[[1,2,3],[1,2,3],1]=0.3
T[[1,2,3],[3,4,5],1]=0.7
```

```python
for s in range(6):
    for sp in range(6):
        if s!=sp:
            R[s,sp,:] = np.abs(s-sp)**(1/3)
        elif s!=0:
            R[s,sp,:] = (s+4)**(-1/2)
        else:
            R[s,sp,:] = 0
```

```python
Q = np.sum(T * (R+((g*np.max(Q, axis=2))[:,:,np.newaxis])), axis=1)[np.newaxis,:,:]
```
