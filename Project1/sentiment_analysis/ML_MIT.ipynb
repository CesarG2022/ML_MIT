```python
def randomization(n):
    """
    Arg:
      n - an integer
    Returns:
      A - a randomly-generated nx1 Numpy array.
    """
    
    A = np.random.random([n,1])
    
    return A
```

```python
def operations(h, w):
    """
    Takes two inputs, h and w, and makes two Numpy arrays A and B of size
    h x w, and returns A, B, and s, the sum of A and B.

    Arg:
      h - an integer describing the height of A and B
      w - an integer describing the width of A and B
    Returns (in this order):
      A - a randomly-generated h x w Numpy array.
      B - a randomly-generated h x w Numpy array.
      s - the sum of A and B.
    """
    A = np.random.random([h,w])
    B = np.random.random([h,w])
    s = A + B

    return A, B, s
```

```python
def norm(A, B):
    """
    Takes two Numpy column arrays, A and B, and returns the L2 norm of their
    sum.

    Arg:
      A - a Numpy array
      B - a Numpy array
    Returns:
      s - the L2 norm of A+B.
    """
    s = np.linalg.norm(A+B)

    return s
```

```python
A = np.array([[1],[2],[3]])
B = np.array([[4],[5],[6]])
norm(A,B)
```

```python
def neural_network(inputs, weights):
    """
     Takes an input vector and runs it through a 1-layer neural network
     with a given weight matrix and returns the output.

     Arg:
       inputs - 2 x 1 NumPy array
       weights - 2 x 1 NumPy array
     Returns (in this order):
       out - a 1 x 1 NumPy array, representing the output of the neural network
    """
    inputs=np.transpose(inputs)
    dot_prod = np.matmul(inputs,weights)
    output = np.tanh(dot_prod)
    
    return output
```

```python
inputs = np.array([[2],[1]])
weights = np.array([[0.3],[0.7]])
neural_network(inputs,weights)
```

## Perceptron without offset

```python
x = np.array([[1,0],[-1,10],[-1,-1]])
y = np.array([-1,1,1])
```

```python
theta = np.array([0,0])

for t in range(10):
  for i in range(3):
    print(f"i: {i}   -   x_i: {x[i]}   -   y_i: {y[i]}   -   theta: {theta}   -   agreement: {y[i]*np.matmul(theta,x[i])} ")
    if y[i]*np.matmul(theta,x[i]) <=0:
      theta += y[i]*x[i]
      
```

## Perceptron with offset

```python
x = np.array([[-4,2],[-2,1],[-1,-1],[2,2],[1,-2]])
y = np.array([1,1,-1,-1,-1])
```

```python
theta = np.array([0,0])
theta_o = 0
i_ini = 0 
missclass=[0,0,0,0,0]
for t in range(10):
  for i in range(5):
    if t==0 and i<i_ini:
      continue
    print(f"i: {i}   -   x_i: {x[i]}   -   y_i: {y[i]}   -   theta: {theta}   -   theta_o:{theta_o}   -   agreement: {y[i] * (np.matmul(theta,x[i]) + theta_o )}")
    if (y[i] * (np.matmul(theta,x[i]) + theta_o )) <= 0 :
      theta += y[i]*x[i]
      theta_o += y[i]
      missclass[i] += 1
print(missclass)
```
